```python
import xarray as xr
import numpy as np
import zarr
```

```python
ds = xr.Dataset({
    'v': (('idx',), np.array([np.datetime64('NaT'), np.datetime64('2010-01-01')], dtype='datetime64[ns]'))
})
ds
```

```python
ds.to_zarr('test.zarr')
```

```python
# same result with mask_and_scale=True or False
from_zarr = xr.open_zarr('test.zarr', mask_and_scale=True).load()
from_zarr
```

```python
from_zarr.to_netcdf('from_zarr.nc', engine='netcdf4')
```

```python
from_nc = xr.open_dataset('from_zarr.nc')
```
